**4. Data Cleaning & Preprocessing**

```python
df.isnull().sum()
```

```python
df['Rating'].fillna(
    df['Rating'].median(),
    inplace=True
)
```

```python
df['Type'].fillna(
    df['Type'].mode()[0],
    inplace=True
)
```

```python
df['Content Rating'].fillna(
    df['Content Rating'].mode()[0],
    inplace=True
)
```

```python
df['Current Ver'].fillna(
    "Unknown",
    inplace=True
)
```

```python
df['Android Ver'].fillna(
    df['Android Ver'].mode()[0],
    inplace=True
)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df['Category'].value_counts()
```

```python
df[df['Category'] == '1.9']
```

```python
df = df[df['Category'] != '1.9']
```

```python
df['Genres'].value_counts()
```

```python
df['Type'].value_counts()
```

```python
df['Content Rating'].value_counts()
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'])
```

```python
df['Installs'] = df['Installs'].str.replace(',', '')
df['Installs'] = df['Installs'].str.replace('+', '')
df['Installs'] = df['Installs'].astype(int)
```

```python
df['Installs'].dtype
```

```python
df['Price'] = df['Price'].str.replace('$', '')
df['Price'] = df['Price'].astype(float)
```

```python
import numpy as np

def convert_size(size):

    if 'M' in str(size):
        return float(size.replace('M', ''))

    elif 'k' in str(size):
        return float(size.replace('k', '')) / 1024

    else:
        return np.nan
```

```python
df['Size'] = df['Size'].apply(convert_size)
```

```python
df['Size'].fillna(
    df['Size'].median(),
    inplace=True
)
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
```

```python
df['Last Updated'].dtype
```

```python
df['Update Year'] = df['Last Updated'].dt.year
df['Update Month'] = df['Last Updated'].dt.month
```
